```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('MortalityDataset.csv')
ax = plt.axes(projection = '3d') #to declare plot axes as 3d

fig = plt.figure(figsize=(18,10))

alive = df[df['MORT']=='alive']
dead = df[df['MORT']=='dead']

ax1 = fig.add_subplot(221)

ax1.scatter(alive['HEIGHT'], alive['WEIGHT'], c='g', label='Alive')
ax1.set_title('3D scatter plot')
ax1.set_xlabel('Height')
ax1.set_ylabel('Weight')
ax1.legend()

ax2 = fig.add_subplot(222)

ax2.scatter(dead['HEIGHT'], dead['WEIGHT'], c='r', label='Dead')
ax2.set_title('3D scatter plot')
ax2.set_xlabel('Height')
ax2.set_ylabel('Weight')
ax2.legend()

ax1 = fig.add_subplot(223,projection='3d')

ax1.scatter(alive['HEIGHT'], alive['WEIGHT'], alive['AGE'], c='g', label='Alive')
ax1.set_title('3D scatter plot based on age for the Living')
ax1.set_xlabel('Height')
ax1.set_ylabel('Weight')
ax1.set_zlabel('Age')
ax1.legend()

ax2 = fig.add_subplot(224,projection='3d')

ax2.scatter(dead['HEIGHT'], dead['WEIGHT'], dead['AGE'], c='r', label='Dead')
ax2.set_title('3D scatter plot based on age for the Dead')
ax2.set_xlabel('Height')
ax2.set_ylabel('Weight')
ax2.set_zlabel('Age')
ax2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(alive['HEIGHT'], alive['WEIGHT'],alive['AGE'], c='green', label='Alive')
ax.scatter(dead['HEIGHT'], dead['WEIGHT'], dead['AGE'], c='red', label='Dead')

ax.set_title('Alive')
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
ax.set_zlabel('Age')
ax.legend()

```

2. 3D BARPLOT

```python
a_df = df[df['BLOOD'] == 'a']
b_df = df[df['BLOOD'] == 'b']
ab_Df = df[df['BLOOD'] == 'ab']
o_df = df[df['BLOOD'] == 'o']


fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111,projection = '3d') #to declare plot axes as 3d

ax.bar(a_df['AGE'],a_df['CHOL'],zs=1,zdir='x',color='red',alpha=0.8, label='A ')
ax.bar(b_df['AGE'],b_df['CHOL'],zs=2,zdir='x',color='blue',alpha=0.8,label='B ')
ax.bar(ab_Df['AGE'],ab_Df['CHOL'],zs=3,zdir='x',color='green',alpha=0.8,label='AB ')
ax.bar(o_df['AGE'],o_df['CHOL'],zs=4,zdir='x',color='purple',alpha=0.8,label='O ')

ax.set_title('Blood Group Chart')
ax.set_ylabel('Age')
ax.set_zlabel('Cholestrol')
ax.legend()
plt.show()
```
